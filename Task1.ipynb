```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_excel('marketting_Campaign.xlsx')#reading excelfile
```

```python
dt=pd.DataFrame(data)#converting it into a dtaframe
df=dt.to_csv('data.csv')#then it converts to csv file
df=pd.read_csv('data.csv')#reading csv file
df
```

```python
print(df.isnull().sum())#printing sum of null value cells
```

```python
df=df.fillna("unknown")#fill the null value cells with unknown
print(df.isnull().sum())#printing to check null value exists
```

```python
df=df.drop_duplicates()#drop duplicates
```

```python
df['Marital_Status']=df['Marital_Status'].str.title().str.strip()
df['Education']=df['Education'].str.title().str.strip()
```

```python
#converting Dt_Customer to date data type
df['Dt_Customer']= pd.to_datetime(df['Dt_Customer'])
print(df['Dt_Customer'])
```

```python
#creating a new column names customer since days it calculates how long they have customer
today=pd.to_datetime('today')
df['Customer_since_years']=(today-df['Dt_Customer']).dt.days//365
df.head()
```

```python
#creating a new column named Age which calculates age of the customer
df['Age']=today.year-df['Year_Birth']
df.head()
```

```python
#replacing columns space with underscores and change to lowercase
df.columns=df.columns.str.lower().str.replace(' ','_')
df.head()
```

```python
#removing duplicated columns
df = df.loc[:, ~df.columns.duplicated()]
```

```python
#reodering some columns for readability 
main_cols = ['year_birth', 'age', 'education', 'marital_status', 'income', 
        'kidhome', 'teenhome', 'dt_customer', 'customer_since_years'] 

remaining_cols=[col for col in df.columns if col not in main_cols]
                
df = df[main_cols+remaining_cols]

df.head()
```

```python
df.to_csv("cleaned_data.csv",index=False)
```
